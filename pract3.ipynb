```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация 1000 случайных чисел из N(0, 1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Построение гистограммы
plt.hist(data, bins=30, edgecolor='black', alpha=0.7, density=True)

# Добавление вертикальной линии в среднем значении
mean = np.mean(data)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Среднее = {mean:.2f}')

# Добавление заголовка и подписей
plt.title('Гистограмма нормального распределения (μ=0, σ=1)')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.legend()

# Отображение графика
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
months = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Авг', 'Сен', 'Окт', 'Нояб', 'Дек']
water_level = [2.3, 2.4, 2.6, 2.8, 3.0, 3.1, 3.2, 3.1, 3.0, 2.8, 2.6, 2.4]

# Создание графика
plt.figure(figsize=(10, 5))
plt.plot(months, water_level, 
         color='blue',
         marker='o',
         linestyle='-',
         linewidth=2,
         markersize=8,
         markerfacecolor='blue',
         markeredgecolor='black',
         label='Уровень воды (м)')

# Настройка заголовка и подписей
plt.title('Уровень воды в озере по месяцам', fontsize=14)
plt.xlabel('Месяцы', fontsize=12)
plt.ylabel('Уровень воды (м)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)

# Отображение графика
plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
seasons = ['Зима', 'Весна', 'Лето', 'Осень']
temperature = [-5, 10, 25, 12]
colors = ['lightblue', 'lightgreen', 'tomato', 'gold']  # Цвета для каждого сезона

# Создание графика
plt.figure(figsize=(8, 5))

# Столбчатая диаграмма с разными цветами
bars = plt.bar(seasons, temperature, color=colors, edgecolor='black', linewidth=1.2)

# Добавление значений температуры над столбцами
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2., height,
             f'{height}°C',
             ha='center', va='bottom', fontsize=10)

# Настройка заголовка и подписей
plt.title('Средняя температура по сезонам', fontsize=14, pad=20)
plt.xlabel('Сезон', fontsize=12)
plt.ylabel('Температура (°C)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Горизонтальная сетка

# Установка диапазона оси Y (для лучшего отображения отрицательных значений)
plt.ylim(min(temperature) - 5, max(temperature) + 5)

plt.tight_layout()  # Автоматическая подгонка layout
plt.show()
```

```python
from scipy.integrate import quad
import numpy as np

# Определяем функцию
def f(x):
    return 1 / x

# Вычисляем интеграл на интервале [1, 5]
result, error = quad(f, 1, 5)

# Выводим результат
print(f"Площадь под функцией на интервале [1, 5]:")
print(f"Точное значение: = {np.log(5):.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем сетку значений x и y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Вычисляем значения функции (конус)
Z = np.sqrt(X**2 + Y**2)

# Создаем 3D-график
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Рисуем поверхность
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

# Добавляем цветовую шкалу
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Настройка подписей (исправленное отображение формулы)
ax.set_title(r'Конус: $f(x, y) = \sqrt{x^2 + y^2}$', fontsize=14)  # Добавлен префикс r
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# Установка угла обзора
ax.view_init(elev=30, azim=45)  # elev - угол наклона, azim - поворот

plt.tight_layout()
plt.show()
```
